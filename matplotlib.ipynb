```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Basic Line Plot")
plt.show()
```

```python
plt.plot(x,y,color = 'red', linestyle = '--', marker ='o')
```
